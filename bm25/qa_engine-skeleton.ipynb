# Preprocessing of questions and reveiws


```python
from nltk import word_tokenize 
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
stop_words=set(stopwords.words('english'))
snowball_stemmer = SnowballStemmer("english")
```

# Questions

```python
QA_DataFrame.drop(['answerTime','answerType','unixTime'],inplace=True,axis=1,errors='ignore')
```

```python
# Tokenization of Questions

filtered_questions = [] 

for questions in QA_DataFrame['question']:
    
    stopwords_removed_questions = []
    stemmed_questions = []
    
    tokens = word_tokenize(questions)
    
    stopwords_removed_questions = [words for words in tokens if not words in stop_words]
    
    
    for words in stopwords_removed_questions:
        stem_word = snowball_stemmer.stem(words)
        stemmed_questions.append(stem_word)
    
    filtered_questions.append(stemmed_questions)
    
```

```python
QA_DataFrame['tokenised_Questions'] = filtered_questions
```

# Reviews

```python
review_DataFrame.drop(['helpful','overall','reviewTime','reviewerID','reviewerName','summary','unixReviewTime'],inplace=True,axis=1,errors='ignore')
```

```python
#review_DataFrame.head()
filtered_df_reviews = review_DataFrame.loc[review_DataFrame['asin'] == 'B000052YM7']
```

```python
# Tokenization of Questions

filtered_reviewes = [] 

for reviews in filtered_df_reviews['reviewText']:
    
    stopwords_removed_reviews = []
    stemmed_reviews = []
    
    tokens = word_tokenize(reviews)
    
    stopwords_removed_reviews = [words for words in tokens if not words in stop_words]
    
    
    for words in stopwords_removed_reviews:
        stem_word = snowball_stemmer.stem(words)
        stemmed_reviews.append(stem_word)
    
    filtered_reviewes.append(stemmed_reviews)
```

```python
filtered_df_reviews['tokenisedReview'] = filtered_reviewes
```

# Handling of Query. 

```python
import nltk
```

```python
user_query = "will this lotion work on my wrinkles?"
stemmed_query = []

# Tokenisation and Stemming 
tokens = word_tokenize(user_query)
for words in tokens:
        stem_word = snowball_stemmer.stem(words)
        stemmed_query.append(stem_word)
        
# Finding the W/H questions        
postag = nltk.pos_tag(stemmed_query)
```

```python
# WH POS Tags - WDT,WP,WP$,WRB
WH_postag = ['WDT','WP','WP$','WRB']
    
WH_Question = [True for value in postag if value[1] in WH_postag]
    
```

```python
# Stop Words Removal
stopwords_removed_query = []

stopwords_removed_query = [words for words in stemmed_query if not words in stop_words]
```
